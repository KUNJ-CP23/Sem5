# Step-13  Perform Condition Selection on DataFrame

```python
# condition mate
df = df[(df['Pclass'] == 1)] 

# age >21 and sex = male
# df[(df['Age'] > 21) & (df['Sex'] == 'male')]
df
```

# Step-14 Compute the sum of value

```python
# ~df[['Fare'] == 1].sum()
print("Sum: ",df[~(df['Fare'] == 1)]['Fare'].sum())
```

# Step-15 Compute the mean of value

```python
print("Mean Value: ",df[~(df['Fare'] == 1)]['Fare'].mean())
```

# Step-16 Count non-null value (column)

```python
print("Total count: ",df[~(df['Fare'] == 1)]['Fare'].count())
```

# Step-17 Find Minimun or Maximum values

```python
max_fare = df['Fare'].max()
print("Maximum Fare:", max_fare)
min_fare = df['Fare'].min()
print("Minimum Fare:", min_fare)

# # for all columns
# print("Max values:\n", df.max())
# print("Min values:\n", df.min())
```
