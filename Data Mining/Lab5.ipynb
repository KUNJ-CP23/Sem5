```python
# fill the avg of mean in age, so take mean of age then fill that mean age in age
data_withfillna = df.copy() #direact age par work nathi karvu
meanAge = df['Age'].mean()
print("Mean Age : ",meanAge)
data_withfillna['Age'] = data_withfillna['Age'].fillna(meanAge)
data_withfillna
```

```python
# fill data with median of ages
data_withfillna = df.copy()
medianAge = df['Age'].median()
print("Median Age : ",medianAge,"years")
data_withfillna['Age'] = data_withfillna['Age'].fillna(medianAge)
data_withfillna
```

```python
# interpolate fills the null values based on upwards and downwards or neighbour values
data_interpolate = df.copy()
data_interpolate['Age']=data_interpolate['Age'].interpolate()
data_interpolate
```

### 3)  Apply Scaling to AGE attribute with min max, decimal scaling and z score.


```python
# Min Max Scaling Formula
# (scaling between range of 0 to 1)
# X-scaled = (X-X_min)/(X_max-X_min)

puredata = df.copy()
puredata['Age'] = puredata['Age'].interpolate()
data_minmax = puredata.copy()
age_mm =data_minmax['Age']
age_min = age_mm.min()
age_max = age_mm.max()
data_minmax['MinMaxAge'] = (age_mm-age_min)/(age_max-age_min)
data_minmax
```

```python
# Decimal Scaling
data_ds = df.copy()
age_ds = data_ds['Age']
age_max = age_ds.abs().max()
temp = len(str(int(age_max)))
data_ds['DecimalScalingAge'] = age_ds/(10**temp)
data_ds[['Age','DecimalScalingAge']]
```

```python
# Z-Score
# X_scaled = (X-mean)/standard_deviation

data_zs = df.copy()
age_zs = data_zs['Age']
age_mean = age_zs.mean()
age_std = age_zs.std()
# print("Mean Age : ",meanAge)
data_zs['ZScoreAge'] = (age_zs-age_mean)/age_std
data_zs
```

```python
# Comparison

com = pd.DataFrame({
    'OriginalAge' : df['Age'],
    'MinMaxAge' : data_minmax['MinMaxAge'],
    'DecimalScalingAge' : data_ds['DecimalScalingAge'],
    'ZScoreAge' : data_zs['ZScoreAge'] 
})

com.head(20)
```
