```python
def euclidean_distance(a, b):
    return np.sqrt(np.sum((np.array(a) - np.array(b))**2))
```

```python
def update_cluster_center(cluster_data):
    sum = [0,0]
    for i in cluster_data:
        sum[0] = sum[0] + i[0]
        sum[1] = sum[1] + i[1]
    return [sum[0]/len(cluster_data), sum[1]/len(cluster_data)]
```

```python

def kmeans(X, k, max_iters=100):
    # Randomly choose centroids
    np.random.seed(42)
    centroids = X[np.random.choice(range(len(X)), k, replace=False)]
    
    for _ in range(max_iters):
        clusters = [[] for _ in range(k)]
        
        # Assign points to nearest centroid
        for point in X:
            distances = [euclidean_distance(point, c) for c in centroids]
            cluster_idx = np.argmin(distances)
            clusters[cluster_idx].append(point)
        
        new_centroids = []
        for cluster in clusters:
            if cluster:  # avoid empty cluster
                new_centroids.append(np.mean(cluster, axis=0))
            else:
                new_centroids.append(centroids[len(new_centroids)])
        
        new_centroids = np.array(new_centroids)
        
        # Convergence check
        if np.allclose(centroids, new_centroids):
            break
        centroids = new_centroids
    
    return centroids, clusters
```

```python
kmeans(X, k=3)  #3 clusters nu example
```

```python
def euclidean_distance(a, b):
    return np.linalg.norm(np.array(a) - np.array(b))
```

```python
def kmedoids(X, k, max_iters=100):
    np.random.seed(42)
    medoids = X[np.random.choice(range(len(X)), k, replace=False)]
    
    for _ in range(max_iters):
        clusters = [[] for _ in range(k)]
        
        # Assign points to nearest medoid
        for point in X:
            distances = np.linalg.norm(medoids - point, axis=1)  # vectorized
            cluster_idx = np.argmin(distances)
            clusters[cluster_idx].append(point)
        
        new_medoids = []
        for cluster in clusters:
            if cluster:
                cluster = np.array(cluster)
                # distance matrix
                dist_matrix = np.linalg.norm(cluster[:, None] - cluster[None, :], axis=2)
                total_distances = dist_matrix.sum(axis=1)
                new_medoids.append(cluster[np.argmin(total_distances)])
            else:
                # keep the old medoid if cluster empty
                new_medoids.append(medoids[len(new_medoids)])
        
        new_medoids = np.array(new_medoids)
        
        if np.allclose(medoids, new_medoids):
            break
        medoids = new_medoids
    
    return medoids, clusters
```
